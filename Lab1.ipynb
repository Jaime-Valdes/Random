```python
import numpy as np
import matplotlib.pyplot as plt

time, position, velocity, acceleration = np.loadtxt("data.csv", skiprows=3, usecols=(0,1,2,3), unpack=True, dtype=float, delimiter=',', max_rows=47)

plt.plot(time, position, label='position (m)')
plt.plot(time, velocity, label='velocity (m/s)')
plt.plot(time, acceleration, label='acceleration (m/s^2)')
plt.legend(loc="upper left", title='legend')
plt.xlabel('Time(s)')


```
